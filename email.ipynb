```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://jobs.nike.com/job/R-40715?from=job%20search%20funnel")
page_data = loader.load().pop().page_content
print(page_data)
print("üìù Length of scraped text:", len(page_data))
```

```python
from langchain_core.prompts import PromptTemplate

# Rebuild the prompt template from scratch to avoid caching issues
prompt_extract = PromptTemplate.from_template("""
### SCRAPED TEXT FROM WEBSITE:
{job_text}

### INSTRUCTION:
From this job listing text, extract job entries and return a JSON array of objects with the following keys:
- "role"
- "experience"
- "skills" (list of technologies, tools, and soft skills based on the role)
- "description"

Return just the JSON array. No explanation or preamble.
""")

# Rename variable to match the template
chain_extract = prompt_extract | llm

# ‚úÖ Pass variable using correct name: "job_text"
res = chain_extract.invoke({"job_text": page_data})

print("üì® LLM Response:", res.content[:500])  # Show preview
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()

try:
    json_res = json_parser.parse(res.content)
    print("‚úÖ Parsed JSON:", json_res)
except Exception as e:
    print("‚ùå JSON Parsing Failed:", e)
    print("üîé Raw Output:", res.content)
```
